##### Minimum spanning tree algorithm

```python
from collections import OrderedDict

class MST:
    def min_spanning_tree(matrix):
        col = []
        result_col = []
        edges = {}
        trees = []

        n = len(matrix)
        for row in matrix:
            for i in range(0, n):    
                for idx, val in enumerate(row):
                    #included negative values
                    #print(i, idx, val)
                    #save edges in format: (row, col): val
                    if idx > i:
                        edges[(i, idx)] = val
                
        sortedEdges = OrderedDict(sorted(edges.items(), key=lambda x: x[1]))
        toBeDeleted = []
        for key, val in sortedEdges.items():

            if val > 0:
                #add into tree:
                #if trees is empty:
                if not trees:
                    smalltree = []
                    smalltree.append(key[0])
                    smalltree.append(key[1])  
                    trees.append(smalltree)
                    #this edge stays
                #if not empty
                else:
                    presence0 = 0
                    presence1 = 0
                    idx_0 = -1
                    idx_1 = -1
                    connects = 0
                    for idx, smalltree in enumerate(trees):
                        if (key[0] in smalltree and presence1 == 1) or (key[1] in smalltree and presence0 == 1):
                            #combine
                            nowidx = idx_0
                            if presence1 == 1:
                                nowidx = idx_1

                            for x in trees[idx]:
                                trees[nowidx].append(x)
                            trees.pop(idx)
                            presence1 = 1
                            presence0 = 1
                            connects = 1
                            break;

                        if key[0] in smalltree and presence0 == 0:
                            #print("first element present.")
                            presence0 = 1
                            idx_0 = idx
                        if key[1] in smalltree and presence1 == 0:
                            #print("second element present.")
                            presence1 = 1
                            idx_1 = idx


                    if presence0 == 0 and presence1 == 0:
                        #print("not present")
                        smalltree = []
                        smalltree.append(key[0])
                        smalltree.append(key[1])
                        trees.append(smalltree)
                        #this edge stays
                    elif presence0 == 0 and presence1 == 1:
                        trees[idx_1].append(key[0])
                    elif presence0 == 1 and presence1 == 0:
                        trees[idx_0].append(key[1])
                    else:
                        if len(trees) == 1 and connects == 0:
                            toBeDeleted.append(key)

        for key in toBeDeleted:
            sortedEdges[key] = -1
        for i in range(0, n):
            row = []
            for j in range(0, n):
                if i < j:
                    row.append(sortedEdges[(i, j)])
                if i == j:
                    row.append(0)
                if i > j:
                    #inverted
                    row.append(sortedEdges[(j, i)])
            result_col.append(row)
        print("\n")
        print("Output matrix: ")
        for row in result_col:
                print(row)  
```

#### Find the shortest path using the breadth-first search algorithm

```python
from collections import deque 
FINAL_ROW = 9
FINAL_COL = 10
# Storing matrix cell cordinates 
class Node:
    def __init__(self, x:int, y:int): 
        self.x = x 
        self.y = y 
# Implementing queue used in BFS
class Queue: 
    def __init__(self, node:Node, dist:int): 
        self.node = node
        self.dist = dist
rowNum = [-1, 0, 0, 1] 
colNum = [0, -1, 1, 0]
# Check if a given cell(row,col) is a valid
def check_valid(row: int, col: int): 
    return (row >= 0) and (row < FINAL_ROW) and (col >= 0) and (col < FINAL_COL) 
# Function to find the shortest path from the start point to the end point  
def shortest_path(matrix, start: Node, end: Node): 
    # check start and end points of the matrix have value 1  
    if matrix[start.x][start.y] !=1 or matrix[end.x][end.y] !=1: 
        return -1
    visited = [[False for i in range(FINAL_COL)] for j in range(FINAL_ROW)] 
    # Mark if visited
    visited[start.x][start.y] = True
    # Create a queue for breadth-first search
    queue = deque() 
    s = Queue(start,0) 
    queue.append(s)
    while queue: 
        curr = queue.popleft()
        node = curr.node 
        if node.x == end.x and node.y == end.y: 
            return curr.dist 
        for i in range(4): 
            row = node.x + rowNum[i] 
            col = node.y + colNum[i] 
            if (check_valid(row,col) and matrix[row][col] == 1 and not visited[row][col]): 
                visited[row][col] = True
                x = Queue(Node(row,col),curr.dist+1) 
                queue.append(x) 
    # If destination is not reachable, return -1
    return -1
            
class SP:
    def find_shortest_path(matrix):
        mtx = matrix
        print("\nPlease enter the position of the initial point")
        start_x = int(input("Enter the row position: "))
        start_y = int(input("Enter the column position: "))
        initial = (start_x, start_y)
        print("Initial point: ", initial)
        print("\nPlease enter the position of the end point")
        end_x = int(input("Enter the row position: "))
        end_y = int(input("Enter the column position: "))
        destination = (end_x, end_y)
        print("Destination point: ", destination)
        init = Node(start_x, start_y) 
        finish = Node(end_x, end_y) 
      
        main = shortest_path(mtx, init, finish) 
      
        if main != -1: 
            print("\nShortest Path is", main) 
        else: 
            print("\nShortest Path does not exist") 
```
